```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("allattractions.csv")
df.info()
```

```python
df.drop_duplicates(subset=["ATT_NAME_TH", "ATT_LOCATION"], keep='first', inplace=True)
df.drop(columns=["ATT_TEL", "ATT_EMAIL", "ATT_WEBSITE", "ATT_FACEBOOK", "ATT_INSTAGRAM", "ATT_LINE", "ATT_TIKTOK", "ATT_YOUTUBE", "ATT_ADDRESS"], inplace=True)
df.info()
```

"ATT_LOCATION" = format(lat, lon)

```python
predf1 = df[["ATT_NAME_TH", "DISTRICT_NAME_TH", "PROVINCE_NAME_TH", "ATT_DETAIL_TH", "ATTR_CATAGORY_TH", "ATTR_SUB_TYPE_TH", "ATT_LOCATION"]].copy()
predf1.info()
```

split "ATT_LOCATION" into ["LATITUDE", "LONGITUDE"]

```python
location = predf1["ATT_LOCATION"].str.split(",", expand=True)
location[0] = location[0].str.strip()
location[1] = location[1].str.strip()

predf1["LATITUDE"] = location[0]
predf1["LONGITUDE"] = location[1]
predf1.drop(columns=["ATT_LOCATION"], inplace=True)
predf1.info()
```

```python
predf1["ATT_DETAIL_TH"] = predf1["ATT_DETAIL_TH"].str.replace("\n", " ").str.replace("\r", " ").str.replace("\t", " ")
predf1.head()
```

```python
# rename columns to simple English
predf1.rename(columns={"ATT_NAME_TH": "name", "DISTRICT_NAME_TH": "district", "PROVINCE_NAME_TH": "province", "ATT_DETAIL_TH": "detail", "ATTR_CATAGORY_TH": "category", "ATTR_SUB_TYPE_TH": "sub_category", "LATITUDE": "latitude", "LONGITUDE": "longitude"}, inplace=True)
predf1.info()
```

```python
predf1["textQuery"] = predf1["name"] + " " + predf1["district"] + " " + predf1["province"] + " ประเทศไทย"
predf1.head()
```

predf1.to_csv("prePlacesData.csv", index_label="idx")  # save with index column named 'idx'
